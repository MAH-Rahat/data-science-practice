```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('animes.csv')
```

```python
print(df.head())
```

```python
df.iloc[[0]]
```

```python
ranked = df['ranked']
score = df['score']
year = df['aired']
members = df['members']
plt.scatter(score, members)
plt.show()
```

```python
df_sort = df.sort_values('score', ascending=False)
print(df_sort.head())
```

```python
max = df['score'].max()
print(max)
mean = df['score'].mean()
print(mean)
```

```python
df.info()
df.value_counts()
```

```python
title = []
score = []

for index, row in df.iterrows():
  if row['score'] > 9:
    score.append(row['score'])
    title.append(row['title'])

plt.bar(title, score, color = 'green')
plt.xlabel('Anime Title')
plt.ylabel('Anime Score')
plt.title('Anime Ratings')
plt.xticks(rotation = 90)
plt.show()
```

```python
min_title = []
min_score = []

for index, row in df.iterrows():
  if row['score'] < 3:
    min_score.append(row['score'])
    min_title.append(row['title'])

plt.bar(min_title, min_score, color = 'red')
plt.xlabel('Anime Title')
plt.ylabel('Anime Score')
plt.title('Anime Low Ratings')
plt.xticks(rotation = 90)
plt.show()
```

```python
title = []
score = []

for index, row in df.iterrows():
  if 'Romance' in row['genre']:
    if row['score'] > 9:
      score.append(row['score'])
      title.append(row['title'])

plt.bar(title, score, color = 'brown')
plt.xlabel('Anime Title')
plt.ylabel('Anime Genre')
plt.title('Comedy Anime')
plt.xticks(rotation = 90)
plt.show()
```
